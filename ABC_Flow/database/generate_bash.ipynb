```python
assert 1 == 2

use_node = 3
njob_node = 24
n_job_pbs = use_node * njob_node
random_order = True

def comm_do_calculate(ini_theta, ini_phi, ini_psi, 
                      ini_x, ini_y, ini_z, 
                      external_flow, table_name, 
                      update_fun, rtol, atol, 
                      ini_t, max_t, eval_dt, save_every, 
                      ABC_A, ABC_B, ABC_C, omega_tail, f, ):
    ABC_info = 'userinput'
    ts = 'python ../../../do_calculate.py  '
    ts = ts + ' -ini_theta %f -ini_phi %f -ini_psi %f' % (ini_theta, ini_phi, ini_psi)
    ts = ts + ' -ini_x %f -ini_y %f -ini_z %f' % (ini_x, ini_y, ini_z)
    ts = ts + ' -external_flow %s -table_name %s' % (external_flow, table_name)
    ts = ts + ' -ini_t %f -max_t %f -eval_dt %e -save_every %d ' % (ini_t, max_t, eval_dt, save_every)
    ts = ts + ' -update_fun %s -rtol %e -atol %e ' % (update_fun, rtol, atol)
    ts = ts + ' -ABC_info %s -ABC_A %f -ABC_B %f -ABC_C %f ' % (ABC_info, ABC_A, ABC_B, ABC_C)
    ts = ts + ' -omega_tail %f -f %s ' % (omega_tail, f)
    return ts

external_flow, table_name = 'ABC', 'ecoB01B05_baseFlow'
update_fun, rtol, atol = '5bs', 1e-9, 1e-12
ini_t, max_t, eval_dt = 0, 10000, 0.0001
ABC_A, ABC_B, ABC_C = 1, 1, 1
omega_tail_list = np.array((0, ))
save_every_list = np.ceil(np.clip(np.log(omega_tail_list), 1, np.inf)) * 10
use_node = 6
njob_node = 23
n_ini = 10
n_job_pbs = 1000000
write_pbs_head = spf.write_pbs_head_q03
name_handle = 'rand_b1'

########################################################################################################
job_dir_handle = os.path.join(table_name, 'A%.2f_B%.2f_C%.2f_w%s_%s' % 
                              (ABC_A, ABC_B, ABC_C, '%.4f', name_handle))
main_run_name = 'main_run_A%.2f_B%.2f_C%.2f_%s.sh' % (ABC_A, ABC_B, ABC_C, name_handle)

# generate all comment file. 
sub_dir_list = []
w_all = []
for omega_tail, save_every in zip(omega_tail_list, save_every_list): 
    job_dir = job_dir_handle % omega_tail
    sub_dir_list.append(job_dir)
    w_all.append(omega_tail)
    
    comm_list = []
    txt_list = []
    for _ in range(n_ini):
        ini_theta, ini_phi, ini_psi = np.random.sample(3) * (1, 2, 2) * np.pi
        ini_x, ini_y, ini_z = np.zeros(3)
        f = 'th%.5f_ph%.5f_ps%.5f_x%.5f_y%.5f_z%.5f' % \
            (ini_theta, ini_phi, ini_psi, ini_x, ini_y, ini_z)
        ts = comm_do_calculate(ini_theta, ini_phi, ini_psi, 
                               ini_x, ini_y, ini_z, 
                               external_flow, table_name, 
                               update_fun, rtol, atol, 
                               ini_t, max_t, eval_dt, save_every, 
                               ABC_A, ABC_B, ABC_C, omega_tail, f)
        comm_list.append(ts)
        txt_list.append(f)
    spf.write_main_run_comm_list(comm_list, txt_list, use_node, njob_node, job_dir, 
                                 write_pbs_head, n_job_pbs, random_order=random_order)


t_name0 = os.path.join(table_name, main_run_name)
with open(t_name0, 'w') as frun:
    frun.write('tdir=$(dirname "$PWD") \n\n')
    for sub_job_dir in sub_dir_list:
        frun.write('sub_dir=$tdir/%s \n' % sub_job_dir)
        frun.write('cd $sub_dir \n')
        frun.write('echo $PWD \n')
        frun.write('bash main_run.sh \n\n')
    frun.write('cd $tdir \n')
```
